#### Replace all odd numbers in arr with -1


```python
# Input
arr = np.arange(10)

# Solution
arr[arr % 2 == 1] = -1

print(arr)
```

#### Convert a 1D array to a 2D array with 2 rows


```python
# Input
arr = np.arange(10)
print(arr)

# Solution
arr = arr.reshape(2, -1)
print(arr)
```

#### Stack arrays a and b vertically and then horizontally


```python
# Input
a = np.arange(10).reshape(2, -1)
b = np.repeat(1, 10).reshape(2, -1)

# Solution
vertical_stack = np.vstack([a, b])
horizontal_stack = np.hstack([a, b])
print(vertical_stack)
print(horizontal_stack)
```

#### Find positions where elements of a and b match


```python
# Input
a = np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6])
b = np.array([7, 2, 10, 2, 7, 4, 9, 4, 9, 8])

# Solution
match = np.where(a == b)
print(match)
```

#### Get the positions of top 5 maximum values in a given array a


```python
# Input
np.random.seed(100)
a = np.random.uniform(1, 50, 20)
print(a)

# Solution:
print(np.sort(a)[::-1][0:5])
```

#### Compute the min-by-max (i.e. min/max) for each row for given 2d numpy array (see [apply_along_axis](https://numpy.org/doc/stable/reference/generated/numpy.apply_along_axis.html#numpy-apply-along-axis))


```python
# Input
np.random.seed(100)
a = np.random.randint(1, 10, [5, 3])
print(a)

# Solution
print(np.apply_along_axis(lambda x: np.min(x) / np.max(x), arr=a, axis=1))
```

#### Find the duplicate entries in the given numpy array and mark them as True. First time occurrences should be False.


```python
# Input
np.random.seed(100)
a = np.random.randint(0, 5, 10)
print(a)

# Solution
# what??

b = np.full_like(a, False)
unique_indices, counts = np.unique(a, return_counts=True)
print(unique_indices, counts)

for e in unique_indices[counts >= 2]:
    b[np.where(a == e)] = True

print(b)
```

#### Find the index of 5th repetition of number 1 in x

Make use of numpy.where


```python
# Input
x = np.array([1, 2, 1, 1, 3, 4, 3, 1, 1, 2, 1, 1, 2])
n = 5
# Soluzione
print(np.where(x == 1)[0][n - 1])
```

#### Compute the moving average of window size 3, for the given 1D array

Make use of numpy.convolve


```python
# Input
arr = np.array([8, 8, 3, 7, 7, 0, 4, 2, 5, 2])

# Solution

moving_avg = np.convolve(arr, np.ones(3) / 3)
# print(moving_avg) # deve ignorare primi ultimi due
print(moving_avg[2:-2])
```

#### Find the mean, median, standard deviation of iris's sepallength (1st column)
